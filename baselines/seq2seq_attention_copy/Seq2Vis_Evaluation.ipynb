# preprocess the cnt, just run one time and no need to repeat

```python
with open('/Users/yuyu/QCRI/Benchmark/Text2Vis/datasets/tvBench_all_good_plus_order.json', 'r') as f:
    data = json.load(f)
```

```python
VQL_dict = dict() #vis_id, vql

for each in data:
    if 'VQL_AST_Order' in each:
        VQL_dict[each['tvBench_id']] = each['VQL_AST_Order']
    elif 'VQL_AST' in each:
        VQL_dict[each['tvBench_id']] = each['VQL_AST']
    else:
        print(each['tvBench_id'])
```

```python
# 去除重复的空格（两个空格连在一起）
for k,v in VQL_dict.items():
    v = v.replace(',',' ,') # 把紧挨着的逗号 空开来
    VQL_dict[k] = " ".join(v.split())
    
```

```python
# CNT of agg---> COUNT

for k, v in VQL_dict.items():
    v_list = v.split(' ')
    agg_y = v_list[v_list.index('FROM')-1]
    if 'CNT' in agg_y.upper():
        agg_y = agg_y.replace('CNT','COUNT')
        v_list[v_list.index('FROM')-1] = agg_y
        VQL_dict[k] = ' '.join(v_list)
        print(VQL_dict[k])
```

```python
# CNT after ORDER BY --> count
for k, v in VQL_dict.items():
    if 'ORDER BY' in v.upper():
        v_list = v.split(' ')
        agg_order = None
        if 'DESC' in v_list:
            agg_order = v_list[v_list.index('DESC')-1]
        elif 'ASC' in v_list:
            agg_order = v_list[v_list.index('ASC')-1]
        else:
            print(k,v)
        if agg_order != None and 'CNT' in agg_order.upper(): 
            agg_order = agg_order.replace('CNT','COUNT')
            try:
                v_list[v_list.index('DESC')-1] = agg_order
            except:
                v_list[v_list.index('ASC')-1] = agg_order

            VQL_dict[k] = ' '.join(v_list)
#             print(VQL_dict[k])
```

# partition VQL into data part (SQL) and vis part

```python
struct_vis_query = dict()

for k, v in VQL_dict.items():
    vis_part = ''
    if 'Visualize LINE' in v:
        vis_part = 'Visualize LINE'
        v = v.replace('Visualize LINE ', '')
    elif 'Visualize BAR' in v:
        vis_part = 'Visualize BAR'
        v = v.replace('Visualize BAR ', '')
    elif 'Visualize PIE' in v:
        vis_part = 'Visualize PIE'
        v = v.replace('Visualize PIE ', '')
    elif 'Visualize SCATTER' in v:
        vis_part = 'Visualize SCATTER'
        v = v.replace('Visualize SCATTER ', '')
    else:
        print('error')
    struct_vis_query[k] = {
        'vis_part': vis_part,
        'data_part':{
            'sql_part': v,
            'binning': ''
        },
        'VQL': v
    }
```

## VIS String Exactly Matching Accuracy

```python
def vis_string_exactly_matching_accuracy(gold_list, predict_list):
    chart_type_abbr = {
        'Bar': "B",
        'Line': "L",
        'Pie': "P",
        'Scatter': 'S',
        'Grouping Line': 'GL',
        'Grouping Scatter': 'GS',
        'Stacked Bar': 'SB'
    }
    vis_string_matching_result = list() # [id, chart_type, hardness, if_match]
    
    for i in range(len(gold_list)):
        this_gold = gold_list[i].replace('\n', '').replace(' ', '')
        this_predict = predict_list[i].replace('\n', '').replace(' ', '')
        
        this_id = str(dev_list_id[i]['tvBench_id']) # get id
        chart_type = chart_type_abbr[property_mapping[this_id]['chart']]
        hardness = property_mapping[this_id]['hardness']
            
        if this_gold == this_predict:
            if_match = 1
        else:
            if_match = 0
            print(this_gold)
            print(this_predict)
            print('\n\n')
            
        vis_string_matching_result.append(
            [this_id, chart_type, hardness, if_match]
        )
        
    df = pd.DataFrame(data = vis_string_matching_result, columns=['id','chart','hardness','if_match'])
    return df
```

```python
df = vis_string_exactly_matching_accuracy(gold_list, predict_list)
```
