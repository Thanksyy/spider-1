## Some dependent functions

```python
chart_type_abbr = {
    'BAR': "B",
    'LINE': "L",
    'PIE': "P",
    'SCATTER': 'S',
    'GROUPING LINE': 'GL',
    'GROUPING SCATTER': 'GS',
    'STACKED BAR': 'SB'
}
```

```python
def inf_chart_type_from_VQL(VQL):
    VQL = VQL.upper()

    V_keywords = VQL.split(' ')
    base_chart = V_keywords[V_keywords.index('VISUALIZE')+1]
    # if two grouping.
    if 'GROUP' in V_keywords and V_keywords.index('GROUP')+3 < len(V_keywords) and V_keywords[V_keywords.index('GROUP')+3] == ',':
        if base_chart == 'BAR':
            return "SB"
        elif base_chart == 'LINE':
            return 'GL'
        elif base_chart == 'SCATTER':
            return 'GS'
        else:
            print('xx')
    # if grouping + binning
    elif 'GROUP BY' in VQL and 'BIN' in VQL:
        if base_chart == 'BAR':
            return "SB"
        elif base_chart == 'LINE':
            return 'GL'
        elif base_chart == 'SCATTER':
            return 'GS'
        else:
            print('yy')
    # if grouping for scatter
    elif base_chart == 'SCATTER' and 'GROUP BY' in VQL:
        return 'GS'
    # grouping for line,  select x, y ... grouping by z
    elif base_chart == 'LINE' and 'GROUP BY' in VQL:
        if V_keywords[V_keywords.index('SELECT')+1] != V_keywords[V_keywords.index('GROUP')+2]:
            return 'GL'
        else:
            return chart_type_abbr[base_chart]
    else:
        return chart_type_abbr[base_chart]
```

## VIS String Exactly Matching Accuracy

```python
def vis_string_exactly_matching_accuracy(gold_list, predict_list):

    vis_string_matching_result = list() # [id, chart_type, hardness, if_match]
    
    for i in range(len(gold_list)):
        this_id = str(dev_list_id[i]['tvBench_id']) # get id
        
        chart_type = chart_type_abbr[property_mapping[this_id]['chart'].upper()]
#         if 'SCATTER' in predict_list[i].replace('\n', '').upper():
#             chart_type = chart_type_abbr[property_mapping[this_id]['chart'].upper()]
#         else:
#             chart_type = inf_chart_type_from_VQL(predict_list[i].replace('\n', ''))
            
        this_gold = gold_list[i].replace('\n', '').replace(' ', '').upper()
        this_predict = predict_list[i].replace('\n', '').replace(' ', '').upper()
        
    
        hardness = property_mapping[this_id]['hardness']
            
        if this_gold == this_predict:
            if_match = 1
        else:
            if_match = 0
            
        vis_string_matching_result.append(
            [this_id, chart_type, hardness, if_match]
        )
        
    df = pd.DataFrame(data = vis_string_matching_result, columns=['id','chart','hardness','if_match'])
    return df
```

## VIS Query Component Matching Accuracy

```python
'''
VQL ----Split to---> each keyowrd part
Identify the keyword indexs in the VQL string.

Visualize Part: B,P,L,S,SB,GL,GS | 
Axis: SELECt [...Axis...] FROM
Data: FROM ... 
'''

def vis_query_component_matching_accuracy(gold_list, predict_list):

    total_grouping, total_join, total_where, total_binning, total_order = 0, 0, 0, 0, 0
    if_match_vis, if_match_axis, if_match_join, if_match_where, if_match_group, if_match_order, if_match_binning = 0, 0, 0, 0, 0, 0, 0

    vis_predict = {
        'B': 0,
        'P': 0,
        'L': 0,
        'S': 0,
        'SB': 0,
        'GL': 0,
        'GS': 0
    }
    
    vis_total = {
        'B': 0,
        'P': 0,
        'L': 0,
        'S': 0,
        'SB': 0,
        'GL': 0,
        'GS': 0
    }
    
    for i in range(len(gold_list)):
        try:
            g_keywords_index = {  # 记录keyword出现的位置
                'FROM': -1,
                'JOIN': -1,
                'WHERE': -1,
                'GROUP': -1,
                'ORDER': -1,
                'BIN': -1
            }
            p_keywords_index = {  # 记录keyword出现的位置
                'FROM': -1,
                'JOIN': -1,
                'WHERE': -1,
                'GROUP': -1,
                'ORDER': -1,
                'BIN': -1
            }

            # VIS Part
            if inf_chart_type_from_VQL(gold_list[i].replace(
                    '\n', '')) == inf_chart_type_from_VQL(
                        predict_list[i].replace('\n', '')):
                if_match_vis += 1  # match

                vis_predict[inf_chart_type_from_VQL(predict_list[i].replace('\n', ''))] += 1
                vis_total[inf_chart_type_from_VQL(gold_list[i].replace('\n', ''))] += 1
            else:
                # non-match
                if_match_vis += 0  
                vis_total[inf_chart_type_from_VQL(gold_list[i].replace('\n', ''))] += 1

            # if two query string exactly matching
            if gold_list[i].replace('\n', '').replace(' ','').upper() == predict_list[i].replace('\n','').replace(' ','').upper():
                total_grouping += 1
                total_join += 1
                total_where += 1
                total_binning += 1
                total_order += 1
                if_match_axis += 1
                if_match_binning += 1
                if_match_group += 1
                if_match_join += 1
                if_match_order += 1
                if_match_where += 1



            else:
                this_id = str(dev_list_id[i]['tvBench_id'])  # get id


                g_keywords, p_keywords = gold_list[i].upper().split(
                    ' '), predict_list[i].upper().split(' ')

                # Axis Part
                if g_keywords[g_keywords.index('SELECT') +
                              1:g_keywords.index('FROM')] == p_keywords[
                                  g_keywords.index('SELECT') +
                                  1:p_keywords.index('FROM')]:
                    if_match_axis += 1
                else:
                    if_match_axis += 0

                for k, v in g_keywords_index.items():
                    try:
                        g_keywords_index[k] = g_keywords.index(k)
                        p_keywords_index[k] = p_keywords.index(k)
                    except:
                        pass

                # Join
                if g_keywords_index['JOIN'] != -1 and p_keywords_index[
                        'JOIN'] != -1:
                    # 都有Join操作
                    total_join += 1


                    if g_keywords_index['WHERE'] != -1 and p_keywords_index[
                            'WHERE'] != -1:
                        if g_keywords[g_keywords.index('JOIN'):g_keywords.index(
                                'WHERE')] == p_keywords[p_keywords.index('JOIN'):
                                                        p_keywords.index('WHERE')]:
                            if_match_join += 1
                        else:
                            if_match_join += 0
                    elif g_keywords_index['GROUP'] != -1 and p_keywords_index[
                            'GROUP'] != -1:
                        if g_keywords[g_keywords.index('JOIN'):g_keywords.index(
                                'GROUP')] == p_keywords[p_keywords.index('JOIN'):
                                                        p_keywords.index('GROUP')]:
                            if_match_join += 1
                        else:
                            if_match_join += 0
                    elif g_keywords_index['ORDER'] != -1 and p_keywords_index[
                            'ORDER'] != -1:
                        if g_keywords[g_keywords.index('JOIN'):g_keywords.index(
                                'ORDER')] == p_keywords[p_keywords.index('JOIN'):
                                                        p_keywords.index('ORDER')]:
                            if_match_join += 1
                        else:
                            if_match_join += 0
                    elif g_keywords_index['BIN'] != -1 and p_keywords_index[
                            'BIN'] != -1:
                        if g_keywords[g_keywords.index('JOIN'):g_keywords.
                                      index('BIN')] == p_keywords[p_keywords.index(
                                          'JOIN'):p_keywords.index('BIN')]:
                            if_match_join += 1
                        else:
                            if_match_join += 0

                    elif g_keywords[g_keywords.index('JOIN'):] == p_keywords[
                            p_keywords.index('JOIN'):]:
                        if_match_join += 1
                    else:
                        if_match_join += 0

                elif g_keywords_index['JOIN'] != -1 and p_keywords_index[
                        'JOIN'] == -1:
                    # ground truth有，而这个没有
                    if_match_join += 0
                    total_join += 1
                else:
                    # 都不存在JOIN 操作
                    pass

                # WHERE
                if g_keywords_index['WHERE'] != -1 and p_keywords_index[
                        'WHERE'] != -1:
                    # 都有where操作
                    total_where += 1
                    if g_keywords_index['GROUP'] != -1 and p_keywords_index[
                            'GROUP'] != -1:
                        if g_keywords[g_keywords.index('WHERE'):g_keywords.index(
                                'GROUP')] == p_keywords[p_keywords.index('WHERE'):
                                                        p_keywords.index('GROUP')]:
                            if_match_where += 1
                        else:
                            if_match_where += 0
                    elif g_keywords_index['ORDER'] != -1 and p_keywords_index[
                            'ORDER'] != -1:
                        if g_keywords[g_keywords.index('WHERE'):g_keywords.index(
                                'ORDER')] == p_keywords[p_keywords.index('WHERE'):
                                                        p_keywords.index('ORDER')]:
                            if_match_where += 1
                        else:
                            if_match_where += 0
                    elif g_keywords_index['BIN'] != -1 and p_keywords_index[
                            'BIN'] != -1:
                        if g_keywords[g_keywords.index('WHERE'):g_keywords.
                                      index('BIN')] == p_keywords[p_keywords.index(
                                          'WHERE'):p_keywords.index('BIN')]:
                            if_match_where += 1
                        else:
                            if_match_where += 0
                    elif g_keywords[g_keywords.index('WHERE'):] == p_keywords[
                            p_keywords.index('WHERE'):]:
                        if_match_where += 1
                    else:
                        if_match_where += 0
                elif g_keywords_index['WHERE'] != -1 and p_keywords_index[
                        'WHERE'] == -1:
                    # ground truth有，而这个没有
                    if_match_where += 0
                    total_where += 1
                else:
                    pass

                # GROUP
                if g_keywords_index['GROUP'] != -1 and p_keywords_index[
                        'GROUP'] != -1:
                    # 都有GROUP操作
                    total_grouping += 1

                    if g_keywords_index['ORDER'] != -1 and p_keywords_index[
                            'ORDER'] != -1:
                        if g_keywords[g_keywords.index('GROUP'):g_keywords.index(
                                'ORDER')] == p_keywords[p_keywords.index('GROUP'):
                                                        p_keywords.index('ORDER')]:
                            if_match_group += 1
                        else:
                            if_match_group += 0
                    elif g_keywords_index['BIN'] != -1 and p_keywords_index[
                            'BIN'] != -1:
                        if g_keywords[g_keywords.index('GROUP'):g_keywords.
                                      index('BIN')] == p_keywords[p_keywords.index(
                                          'GROUP'):p_keywords.index('BIN')]:
                            if_match_group += 1
                        else:
                            if_match_group += 0
                    elif g_keywords[g_keywords.index('GROUP'):] == p_keywords[
                            p_keywords.index('GROUP'):]:
                        if_match_group += 1
                    else:
                        if_match_group += 0
                elif g_keywords_index['GROUP'] != -1 and p_keywords_index[
                        'GROUP'] == -1:
                    # ground truth有，而这个没有
                    if_match_group += 0
                    total_grouping += 1
                else:
                    pass

                # ORDER
                if g_keywords_index['ORDER'] != -1 and p_keywords_index[
                        'ORDER'] != -1:
                    # 都有GROUP操作
                    total_order += 1

                    if g_keywords_index['BIN'] != -1 and p_keywords_index[
                            'BIN'] != -1:
                        if g_keywords[g_keywords.index('ORDER'):g_keywords.
                                      index('BIN')] == p_keywords[p_keywords.index(
                                          'ORDER'):p_keywords.index('BIN')]:
                            if_match_order += 1
                        else:
                            if_match_order += 0
                    elif g_keywords[g_keywords.index('ORDER'):] == p_keywords[
                            p_keywords.index('ORDER'):]:
                        if_match_order += 1
                    else:
                        if_match_order += 0
                elif g_keywords_index['ORDER'] != -1 and p_keywords_index[
                        'ORDER'] == -1:
                    # ground truth有，而这个没有
                    if_match_order += 0
                    total_order += 1
                else:
                    pass

                # BIN
                if g_keywords_index['BIN'] != -1 and p_keywords_index['BIN'] != -1:
                    # 都有GROUP操作
                    total_binning += 1

                    if g_keywords[g_keywords.index('BIN'):] == p_keywords[
                            p_keywords.index('BIN'):]:
                        if_match_binning += 1
                    else:
                        if_match_binning += 0
                elif g_keywords_index['BIN'] != -1 and p_keywords_index[
                        'BIN'] == -1:
                    # ground truth有，而这个没有
                    if_match_binning += 0
                    total_binning += 1
                else:
                    pass
    
        except:
            print('errors~')

    
    for v_type, value in vis_predict.items():
        print(v_type, round(value/vis_total[v_type]*100, 2))

        
    print('\n')    
    print('Vis: ', round(if_match_vis / len(gold_list)*100, 2), '\n'
          'Axis: ', round(if_match_axis / len(gold_list)*100, 2), '\n'
          'Where: ', round(if_match_where / (total_where)*100, 2), '\n'
          'Join: ', round(if_match_join / (total_join)*100, 2), '\n'
          'Grouping: ', round(if_match_group / (total_grouping)*100, 2), '\n'
          'Binning: ', round(if_match_binning / (total_binning)*100, 2), '\n'
          'Order: ', round(if_match_order / (total_order)*100, 2), '\n')
    
    
    for v_type, value in vis_predict.items():
        print(round(value/vis_total[v_type]*100, 1) , '&')

       
    print( round(if_match_vis / len(gold_list)*100, 1), '&',
           round(if_match_axis / len(gold_list)*100, 1), '&',
           round(if_match_where / (total_where)*100, 1), '&',
           round(if_match_join / (total_join)*100, 1), '&',
           round(if_match_group / (total_grouping)*100, 1), '&',
           round(if_match_binning / (total_binning)*100, 1), '&',
           round(if_match_order / (total_order)*100, 1))
    
```
